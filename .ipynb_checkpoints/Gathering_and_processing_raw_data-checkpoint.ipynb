### Gathering the Finnish costs of building data

```python
url='https://pxdata.stat.fi:443/PxWeb/api/v1/en/StatFin/rki/statfin_rki_pxt_11na.px'
```

```python
query = {
  "query": [
    {
      "code": "Perusvuosi",
      "selection": {
        "filter": "item",
        "values": [
          "1990_100"
        ]
      }
    },
    {
      "code": "Indeksi",
      "selection": {
        "filter": "item",
        "values": [
          "Kokonaisindeksi",
          "Työpanokset",
          "Tarvikepanokset",
          "Palvelut"
        ]
      }
    },
    {
      "code": "Tiedot",
      "selection": {
        "filter": "item",
        "values": [
          "pisteluku",
          "vuosimuutos"
        ]
      }
    }
  ],
  "response": {
    "format": "json-stat2"
  }
}
```

```python
session = requests.Session()

response = session.post(url, json=query)
response_json = json.loads(response.content.decode('utf-8-sig'))

response_json
```

```python
colnames = ["Total", "Labour", "Materials", "Services"]
```

```python
response_json['value']
```

```python
2022-1990
```

```python
response_json['value']
```

```python
year = []
total = []
labour = []
materials = []
services = []
for x in range(0, len(response_json['value']), 8):
    total.append(response_json['value'][x])
    labour.append(response_json['value'][x+2])
    materials.append(response_json['value'][x+4])
    services.append(response_json['value'][x+6])
    year.append(int(x/8+1990))
```

```python
finnish_costs = pd.DataFrame(list(zip(total,labour,materials,services)), index = year, columns =[colnames])
```

```python
finnish_costs
```

### Reading in the social housing completion data

```python
# create an Empty DataFrame object
social_housing_ireland = pd.read_csv(csv_files[0], encoding = 'cp1252', usecols=[2,4])
social_housing_ireland.columns = ["Local Authority", "No. of Units"]
social_housing_ireland["Quarter"] = csv_files[0][-11:-4]

# loop over the list of csv files
for f in csv_files[1:]:
      
    # read the csv file
    df = pd.read_csv(f, encoding = 'cp1252', usecols=[2,4])
    df.columns = ["Local Authority", "No. of Units"]
    df["Quarter"] = f[-11:-4]
    
    social_housing_ireland = pd.concat([social_housing_ireland, df])
```

### Reading in the Irish house construction cost index data

```python
path = os.getcwd() + "/RawData/Irish Datasets/"
irish_costs = pd.read_csv(path + "HSM09 — House Construction Cost Index.csv")
```

```python
irish_costs.head()
```

```python
#Check for value in the columns
print(irish_costs["STATISTIC"].nunique())
print(irish_costs["STATISTIC Label"].nunique())
print(irish_costs["TLIST(M1)"].nunique())
print(irish_costs["Month"].nunique())
print(irish_costs["C02196V02652"].nunique())
print(irish_costs["State"].nunique())
print(irish_costs["UNIT"].nunique())
```

```python
irish_costs.rename(columns = {"VALUE": "House Construction Cost Index"}, inplace= True)
```

```python
irish_costs = irish_costs.iloc[:,[3, 6, 7]]
```

```python
irish_costs.head()
```

```python
irish_costs.tail()
```

```python
#Create a Quarter variable in same format as completion data dataframe
irish_costs["Quarter"] = " "
for i in range(0,len(irish_costs.Month)):
    year=irish_costs["Month"][i][0:4]
    month=irish_costs.Month[i][5:7]
    irish_costs["Month"][i] = f"{year}{month}"
    irish_costs.Month[i] = irish_costs.Month[i].replace("O", "0")
    try:
        if month == '01' or month == '02' or month == '03':
            irish_costs["Quarter"][i] = (f"q1_{year}")
        elif month == '04' or month == '05' or month == '06':
            irish_costs["Quarter"][i] = (f"q2_{year}")
        elif month == '07' or month == '08' or month == '09':
            irish_costs["Quarter"][i] = (f"q3_{year}")
        elif month == '10' or month == '11' or month == '12':
            irish_costs["Quarter"][i] = (f"q4_{year}")
    except:
        print("Month format does not match.")
```

```python
irish_costs.head()
```

### Checking for missing date values

```python
start_dt = date(1975, 1, 1)
end_dt = date(2017, 8, 1)

# difference between current and previous date
#delta = timedelta(month=1)

# store the dates between two dates in a list
dates = []

while start_dt <= end_dt:
    # add current date to list by converting  it to iso format
    dates.append(start_dt.isoformat())
    # increment start date by timedelta
    start_dt += relativedelta(months=1)
    
dates_df = pd.DataFrame(dates, columns = ["date"])

#print(dates)
```

```python
#Make the Month variable a datetime object
for i in range(0,len(irish_costs.Month)):
    irish_costs.Month[i] = pd.to_datetime(irish_costs.Month[i],format='%Y%m')
```

```python
f = plt.figure()
ax1 = f.add_subplot(121)
ax1.set_title("No differencing")
ax1.plot(irish_costs["House Construction Cost Index"])

ax2 = f.add_subplot(122)
plot_acf(irish_costs["House Construction Cost Index"], ax=ax2)

plt.show
```

```python
f = plt.figure()
ax1 = f.add_subplot(121)
ax1.set_title("First order differencing")
ax1.plot(irish_costs["House Construction Cost Index"].diff())

ax2 = f.add_subplot(122)
plot_acf(irish_costs["House Construction Cost Index"].diff(), ax=ax2)

plt.show
```

```python
#No differencing
result = adfuller(irish_costs["House Construction Cost Index"])

print('ADF Statistic: {}'.format(result[0]))
print('p-value: {}'.format(result[1]))
print('Critical Values:')
for key, value in result[4].items():
    print('\t{}: {}'.format(key, value))
```

```python
# 1st order differencing

result = adfuller(irish_costs["House Construction Cost Index"].diff().dropna())

print('ADF Statistic: {}'.format(result[0]))
print('p-value: {}'.format(result[1]))
print('Critical Values:')
for key, value in result[4].items():
    print('\t{}: {}'.format(key, value))
```

### Fit ARIMA Model

```python
model = ARIMA(irish_costs["House Construction Cost Index"], order=(50,2,0))
model_fit = model.fit()
predictions = model.fit().predict()
```

```python
# summary of fit model
print(model_fit.summary())

# line plot of residuals
residuals = DataFrame(model_fit.resid)
residuals.plot()
plt.show()

#density plot of residuals
residuals.plot(kind='kde')
plt.show()

#summary stats of residuals showing mean ~= 0.
print(residuals.describe())
```

```python
tscv = TimeSeriesSplit(n_splits = 10)
rmse = []
for train_index, test_index in tscv.split(irish_costs_for_ARIMA):
    cv_train, cv_test = irish_costs_for_ARIMA.iloc[train_index], irish_costs_for_ARIMA.iloc[test_index]
    model = ARIMA(cv_train["House Construction Cost Index"], order=(50, 2, 0)).fit()
    predictions = model.predict(cv_test.index.values[0], cv_test.index.values[-1])
    true_values = cv_test["House Construction Cost Index"]
    rmse.append(np.sqrt(mean_squared_error(true_values, predictions)))
```

```python
# split into train and test sets
X = irish_costs_for_ARIMA["House Construction Cost Index"]
size = int(len(X) * 0.66)
train, test = X[0:size], X[size:len(X)]
history = [x for x in train]
predictions = list()
```

```python
model_test = ARIMA(history, order=(50,2,0))
model_test_fit = model_test.fit()
```

```python
output_test = model_test_fit.forecast(len(test))
output_test
```

```python
# walk-forward validation
for t in range(len(test)):
    yhat = output_test[t]
    predictions.append(yhat)
    obs = test[test.index[t]]
    history.append(obs)
    print('predicted=%f, expected=%f' % (yhat, obs))
```
